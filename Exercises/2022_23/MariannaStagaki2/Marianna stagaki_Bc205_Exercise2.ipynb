Παρακάτω βρίσκεται ο κώδικας για την δημιουργία του Position Weight Matrix

```python
#creation of pwm
def take_seqs(file):
    with open(file, "r+") as file:
            seqs = file.readlines()
            sequences = [i.replace("\n", "") for i in seqs]
            return sequences
    
def pwm(sequences):
    nuc = ['A', 'C', 'G', 'T']
    profile=[[0 for i in range(len(sequences[0]))] for j in range(len(nuc))]
    for instance in sequences:
        for j in range(len(instance)):
            residue=instance[j]
            profile[nuc.index(residue)][j]+=1
            profile[nuc.index(residue)][j]=float(profile[nuc.index(residue)][j])
    import numpy as np
    pwm = np.array(profile)
    pwm = pwm/len(sequences)
    return(pwm)

pwm(take_seqs("motifs_in_sequence.fa"))
```

Παρακάτω παράγονται τα τυχαία μοτίβα, χρησιμοποιώντας την random randrange εντολή της python.

```python
import random
import math
import numpy as np

def generate_random_motifs(file, k):
    with open(file, "r+") as file:
        seqs = file.readlines()
        seqs2 = [i.replace("\n", "") for i in seqs]
    possible_motifs = []
    for seq in seqs2:
        first_index = random.randrange(0,len(seq)-k)
        possible_motifs.append(seq[first_index:first_index+k])
    return possible_motifs

generate_random_motifs("motifs_in_sequence.fa", 5)[:10]
### generate_random_motifs is a list
pwm(generate_random_motifs("motifs_in_sequence.fa", 5))
### pwm is an numpy array
```

Παρακάτω υπολογίζεται με την μέθοδο του Shannon για την εντροπία, η πληροφορία που παίρνουμε για κάθε PWM και ζητάμε να επιστρέψει το PWM με πληροφορία μεγαλύτερη από ένα threshold, το οποίο μπορεί να είναι μέχρι 2 (το μέγιστο). Το threshold μπορούμε να το ζητήσουμε πχ μεγαλύτερο από 1,6. Σίγουρα υπάρχει κάποιο λάθος ή πολύ πιθανό να μην κατάλαβα κάτι καλά στην υλοποιήση του κώδικα διότι οι τιμές που βγαίνουν για κάθε PWM είναι πολύ μικρότερες από το αναμενόμενο και δεν καταλήγει ποτέ ο κώδικας σε αποτέλεσμα.

```python
def de_novo_motif(file,k, threshold):
    score = 0
    scores = []
    info = 0 
    while info < threshold:
        random_motifs = generate_random_motifs(file, k)
        for count in np.transpose(pwm(random_motifs)):
                score = 0 
                for frequency in count:
                    if frequency > 0:
                        score -= frequency * np.log2(frequency)
                scores.append(2-score)
        info =  sum(scores)/len(scores)
        print(info)
        if info >= threshold:
            return pwm(random_motifs)
de_novo_motif(("motifs_in_sequence.fa"),5, 1.6)
```
