#### 2. Finding palindromes in a sequence using recursion
There is an easy way to define a palindrome sequence _S_ of length _l_ if we consider its indices _1:l_. This is:
```
S is a palindrome if S[1:l]==S[l:1]
```
* Write a function **using recursion** that will examine:
a. whether a sequence contains a palindrome in the middle (or not).
b. will return the longest possible palindrome 


````python
# Function to read a fasta file sequence

def readfasta(fastafile):
    import regex as re
    f=open(fastafile, 'r')
    seq = ""
    total = 0
    for line in f:
        x=re.match(">", line)
        if x == None:
            length=len(line)
            total=total+length
            seq=seq+line[0:length-1]
    f.close()
    return(seq)

ecoliseq=readfasta("files/ecoli.fa")
print(ecoliseq[1:100])
````

````python
# function to find fixed elements in string
def findoccurrence(sequence, pattern):
    position=0
    occurrences=list()
    while(len(sequence)>len(pattern)):
        i=sequence.index(pattern)+1
        position=position+i-1
        occurrences.append(position)
        sequence=sequence[i:]
    return(occurrences)

findoccurrence("ATATTTAAAAATTTT", "A")

"""
sequence="ATATTTAAAAATTTT"
pattern="TTT"
position=0
occurrences=list()
while(len(sequence)>len(pattern)):
    i=sequence.index(pattern)+1
    position=position+i-1
    occurrences.append(position)
    sequence=sequence[i:]
print(occurrences)
"""
````

````python

# Function for palindrome search

string_for_function = str(ecoliseq[1:50])
print(string_for_function)

string_for_function = string_for_function.lower()
string_for_function = ''.join(i for i in string_for_function if i.isalnum())
print(string_for_function)
````

````python
string_for_function = "tttttttcgaccaaaggtaacgaggtaacaaccatgcgagtgttgaagta"
print(len(string_for_function))
print(string_for_function)
def is_palindrome(s):
  return s == s[::-1]

def palindrome_search(string):
  if is_palindrome(string):
    return string

  x = palindrome_search(string[:-1])
  y = palindrome_search(string[1:])

  if len(x) > len(y):
    return x
  else:
    return y

print(palindrome_search(string_for_function))
````
